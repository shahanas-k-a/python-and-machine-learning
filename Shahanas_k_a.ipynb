we generally start our code by importing the libraries which we will use throughout the program.

```python
import numpy as np 
import pandas as pd 
import sklearn 
```

Now we will load our data that we use to train our model 

Here we are using boston house pricing datasets that are available in sklearn library itself 

```python
from sklearn.datasets import load_boston 
df = load_boston() 
```

```python
df.keys() #Return all keys of the dataset dictionary 
```

```python
print(df.DESCR) #Info about the dataset 
```

we are converting the datasets to panda.

```python
boston =pd.DataFrame(df.data,columns=df.feature_names) 
boston.head()
```

Adding a new coloumn of target values to the dataframe 

```python
boston['MEDV'] =df.target 
boston.head() 
```

Check the dataset contain any null value or not 

```python
boston.isnull()
```

isnull return true or false for each of the cell in the dataframe.
we can use sum() functionto count all the cells with true value 


```python
boston.isnull().sum()
```

We never train the model on all the data that we have, we always make sure to atleast have test dataset,which is different from the training dataset 

```python
from sklearn.model_selection import train_test_split 
X = boston.drop('MEDV' , axis=1) 
Y = boston['MEDV'] 
X_train, X_test, Y_train,  Y_test = train_test_split(X, Y, test_size =0.15, random_state=5) 
print(X_train.shape) 
print(X_test.shape) 
print(Y_train.shape) 
print(Y_test.shape) 
```

Now let's we import the linear regression model from sklearn and train it to thetraining dataset 

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error 
```

```python
lin_model = LinearRegression() 

lin_model.fit(X_train ,Y_train)
```

```python
y_train_predict = lin_model.predict(X_train) 
rmse = (np.sqrt(mean_squared_error(Y_train ,y_train_predict))) 

print("The model performance for training set") 
print(' RMSE is {} '.format(rmse)) 
print("\n") 

#on testing set 
y_test_predict = lin_model.predict(X_test) 
rmse = (np.sqrt(mean_squared_error(Y_test ,y_test_predict))) 

print("The model performance for testing set") 
print(' RMSE is {} '.format(rmse)) 
print("\n") 
```
